# Matplotlib Application simple

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 100)
y =  x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0 , 1, 1])
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("X vs y")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0, 1, 1])
ax2 = fig.add_axes([0.2,.5,.2,.2])

ax1.plot(x,y, color='c')
ax2.plot(x,y, color='r')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax.plot(x,z)
ax2.plot(x, y)
ax2.set_xlim([20, 22])
ax2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].plot(x, y, color='blue', ls='--', lw=3)
axes[1].plot(x,z, color='r', lw=3)
```

```python
data = dict(type='choropleth',
           locations=gdp['CODE'],
           z = gdp['GDP (BILLIONS)'],
           text = gdp['COUNTRY'],
           colorbar=dict(title='GDP in Billions USD'))


layout = dict(title='2014 Global GDP',
             geo=dict(showframe=False,
                     projection={'type':'mercator'}))
```

```python
choromap3 = go.Figure(data=[data], layout=layout)
iplot(choromap3)
```
